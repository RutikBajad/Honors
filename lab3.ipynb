```python
mean = df.groupby("City")["Temperature"].mean()
median = df.groupby("City")["Temperature"].median()
mode = df.groupby("City")["Temperature"].apply(lambda x: x.mode().tolist())

print("\nMean:")
print(mean)
print("\nMedian:")
print(median)
print("\nMode:")
print(mode)
```
